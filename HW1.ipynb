```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/births/US_births_2000-2014_SSA.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.mean(df['births'])
plt.scatter(df['day_of_week'], df['births'])
plt.xlabel('day of week')
plt.ylabel('# of births');
```

```python
grp = df.groupby('date_of_month')
birth_by_date = grp.mean()

plt.plot(birth_by_date.index, birth_by_date.births)
plt.xlabel('date of month')
plt.ylabel('avg. births per day');
```

```python
year_grp = df.groupby('year')
year_grp.sum()
```

```python
births_by_year = year_grp.sum()

plt.plot(births_by_year.index, births_by_year.births)
plt.ylabel('# of births')
plt.xlabel('year')
plt.axvline(2008)
#2008 was the US recession, we see a dramatic drop in births after this year, meaning there is a large correlation between economy and number of births
```

```python
first_day_sel = (df['date_of_month'] == 1) | (df['date_of_month'] == 2)

first_day_grp = df[first_day_sel].groupby('month')
first_day_birth_by_month = first_day_grp.mean()

plt.plot(first_day_birth_by_month.index, first_day_birth_by_month.births)
plt.xlabel('month')
plt.ylabel('avg. births per day');

#here we can see that people do not like to have children the first day of the month apparently, and tend to have them later on
```
